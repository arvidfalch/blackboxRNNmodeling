# Impulse response test

This notebook loads the audio from the impulse response test of the effects and plots them to inspect their behaviour. 


```python
dry, dummy = librosa.load('./MLDataset2/pulse-Dry.wav', sr=sr)
SD1, dummy = librosa.load('./MLDataset2/pulse-SD1.wav', sr=sr)
CE5, dummy = librosa.load('./MLDataset2/pulse-CE5.wav', sr=sr)
B65, dummy = librosa.load('./MLDataset2/pulse-B65.wav', sr=sr)
```

```python
# The full audio files visualized

plt.figure(figsize=(16,6))
plt.plot(dry)
plt.plot(SD1)
plt.plot(CE5)
plt.plot(B65)
plt.legend(['Dry','SD1', 'CE5', 'B65'])
plt.show()
```
